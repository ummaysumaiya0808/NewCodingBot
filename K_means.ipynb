**Optimizing K-Means.**

```python
km=KMeans(n_clusters=8)

wcss=[]
clusters_range= range(1,15)
for k in clusters_range:
  km2=KMeans(n_clusters=k)
  km2.fit(df[['income', 'score']])
  wcss.append(km2.inertia_)
```

```python
print
```

```python
wcss
```

```python
plt.plot(clusters_range, wcss, marker='o')
plt.xlabel("Clusters")
```

After Optimizing the clusters.

```python
km3=KMeans(n_clusters=5) #Creating K-means object.
km3.fit(df[['score','income']]) #---. Training ML model
```

```python
df['optimize_clusters']= km3.predict(df[['score','income']])
df.head()
```

```python
df0=df[df.optimize_clusters==0]
df1=df[df.optimize_clusters==1]
df2=df[df.optimize_clusters==2]
df3=df[df.optimize_clusters==3]
df4=df[df.optimize_clusters==4]

plt.figure(figsize=(10,7.5))

plt.scatter(df0.score, df0.income, color='black', label='Cluster 0')
plt.scatter(df1.score, df1.income, color='red', label='Cluster 1')
plt.scatter(df2.score, df2.income, color='magenta', label='Cluster 2')
plt.scatter(df3.score, df3.income, color='green', label='Cluster 3')
plt.scatter(df4.score, df4.income, color='pink', label='Cluster 4')

plt.xlabel('Score in (1-100)')
plt.ylabel('Income')
plt.title('Score nd income plot')

plt.legend()
plt.show()
```
