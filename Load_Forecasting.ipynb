```python
import pandas as pd
import matplotlib.pyplot as plt
from pycaret.regression import *
```

```python
df = pd.read_csv('/content/energy_dataset.csv')
df
```

Here, we have eliminated all 0/blank, and null columns.

```python
df=df.drop('Unnamed: 4', axis=1)
df.head()
```

```python
# Plot the actual load and forecasted load
plt.figure(figsize=(13, 6))
plt.plot(df['total load forecast'], label='Actual Load', marker='o')
plt.plot(df['total load forecast'], label='Forecasted Load', marker='+')
plt.title('Actual vs Forecasted Load')
plt.xlabel('Time')
plt.ylabel('Load')
plt.legend()
plt.show()
```
